```python
from pyspark.sql.functions import *
```

```python
stor_silver_df = spark.read.table("retail_analytics.bronze.stores")
```

```python
stor_silver_df = (
    spark.read.table("retail_analytics.bronze.stores")
    .dropDuplicates(["Store_ID"])
    .withColumn("Country", trim(col("Country")))
    .withColumn("City", trim(col("City")))
    .withColumn("Store_Name", trim(col("Store_Name")))
    .withColumn("ZIP_Code", trim(col("ZIP_Code")))
    .withColumn("ingestion_date", to_date(col("ingestion_ts")))
    .fillna({
        "Country": "Not available",
        "City": "Not available",
        "Store_Name": "Not available",
        "ZIP_Code": "Not available"
    })
)
```

```python
(stor_silver_df.write
 .format("delta")
 .mode("overwrite")
 .saveAsTable("retail_analytics.silver.stores")
)
```
