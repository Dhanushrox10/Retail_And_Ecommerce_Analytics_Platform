```python
from pyspark.sql.functions import *
```

```python
emp_silver_df = spark.read.table("retail_analytics.bronze.employees")
```

```python
emp_silver_df = (
    spark.read.table("retail_analytics.bronze.employees")
    .dropDuplicates(["Employee_ID"])
    .withColumn("Name", initcap(trim(col("Name"))))
    .withColumn("Position", initcap(trim(col("Position"))))
    .withColumn("ingestion_date", to_date(col("ingestion_ts")))
    .fillna({
        "Name": "Not available",
        "Position": "Not available"
    })
)
```

```python
(emp_silver_df.write
 .format("delta")
 .mode("overwrite")
 .saveAsTable("retail_analytics.silver.employees")
)
```
