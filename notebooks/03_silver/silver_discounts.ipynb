###Importing Pyspark Functions

```python
from pyspark.sql.functions import *
```

###Reading data from bronze file.

```python
disc_silver_df = spark.read.table("retail_analytics.bronze.discounts")
```

###Basic cleaning process and handling null values

```python
disc_silver_df = (
    spark.read.table("retail_analytics.bronze.discounts")
    .dropDuplicates()
    .withColumnRenamed("Discont", "Discount")
    .withColumn("Description", trim(col("Description")))
    .withColumn("Category", trim(col("Category")))
    .withColumn("Sub_Category", trim(col("Sub_Category")))
    .withColumn("ingestion_date", to_date(col("ingestion_ts")))
    .fillna({
        "Description": "Not available",
        "Category": "Not available",
        "Sub_Category": "Not available"
    })
)
```

###Saving the table

```python
(disc_silver_df.write
 .format("delta")
 .mode("overwrite")
 .saveAsTable("retail_analytics.silver.discounts")
)
```
